```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
df = pd.read_csv("heights.csv")
df
```

```python
plt.hist(df.height, bins=20, rwidth=0.8,density=True)
plt.xlabel('Height (inches)')
plt.ylabel('Count')

x = np.arange(df.height.min(), df.height.max(), 0.1)
plt.plot(x, norm.pdf(x, df.height.mean(), df.height.std()))
```

### std

```python
upper_limit = df.height.mean() + 3*df.height.std()
upper_limit
```

```python
lower_limit = df.height.mean() -3*df.height.std()
lower_limit
```

```python
df[(df.height > upper_limit) | (df.height < lower_limit)]
```

```python
df2 = df[(df.height < upper_limit) & (df.height > lower_limit)]
df2
```

### z - score

```python
df['z_score'] = (df.height - df.height.mean()) / df.height.std()
df
```

```python
df[(df.z_score < -3) | (df.z_score > 3)]
```

```python
df3 = df[(df.z_score >- 3) & (df.z_score < 3)]
df3
```
