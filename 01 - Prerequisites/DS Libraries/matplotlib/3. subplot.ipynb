```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#
x = np.linspace(-10,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = x**3

# 
plt.subplot(2,2,1) ; plt.plot(x,y1,'r',label = 'Sin') ; plt.legend(loc=1)

plt.subplot(2,2,2) ; plt.plot(x,y2,'g',label = 'Cos') ; plt.legend(loc=2)

plt.subplot(2,2,3) ; plt.plot(x,y3,'b',label = 'x**2') ; plt.legend(loc=3)

plt.subplot(2,2,4) ; plt.plot(x,y4,'y',label = 'x**3') ; plt.legend(loc=4)
```
