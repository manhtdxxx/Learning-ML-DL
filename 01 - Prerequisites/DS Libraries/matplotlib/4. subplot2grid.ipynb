```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#
x = np.linspace(-10,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = x**3

# chia 3 hàng 3 cột, có thứ tự từ 0 đến 2
plt.subplot2grid((3,3),(0,0),rowspan=2,colspan=2) ; plt.plot(x,y1,'r:',label = 'Sin') ; plt.legend(loc=1)
plt.subplot2grid((3,3),(2,0),colspan=2) ; plt.plot(x,y2,'g-.',label = 'Cos') ; plt.legend(loc=2)
plt.subplot2grid((3,3),(0,2)) ; plt.plot(x,y3,'y--',label = 'x**2') ; plt.legend(loc=3)
plt.subplot2grid((3,3),(1,2),rowspan=2) ; plt.plot(x,y4,'k.',label = 'x**3') ; plt.legend(loc=4)
```
