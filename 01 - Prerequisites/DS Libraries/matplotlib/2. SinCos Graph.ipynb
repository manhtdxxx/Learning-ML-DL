```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# CHUẨN BỊ DATA
x = np.linspace(0,20,1000)
y = np.sin(x)
#print(y)

# data visualization
plt.plot(x,y,label = 'Đồ thị hình sin')

#
plt.title('Sóng siêu âm')
plt.xlabel('Thời gian')
plt.ylabel('Biên độ')

# turn on grid
plt.grid(True)

# coordinate and content of figure text
plt.figtext(0.3,-0.05,'Hình 1: Sóng siêu âm ở tần số 1MHz')

# location of legend in range of 0 - 10
plt.legend(loc=1) # chú thích trong biểu đồ

# save figure
plt.savefig('dothisin_songsieuam.png')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# CHUẨN BỊ DATA
x = np.linspace(0,20,1000)
y1 = np.sin(x)
y2 = np.cos(x)

# data visualization
plt.plot(x,y1,'g:',label = 'Sin') # b-blue, g-green, r-red, y-yellow
plt.plot(x,y2,'r-.',label = 'Cos') # : , -- , -. , * , + , - , . , o , x , s-square , h-hexagon , d-diamond
plt.legend(loc=4)

# điều chỉnh trục x,y
plt.axis([-5,25,-2,2]) # hoặc plt.xlim ([... ,...]) & plt.ylim([])
plt.axis('tight')
plt.axis('equal')
```
