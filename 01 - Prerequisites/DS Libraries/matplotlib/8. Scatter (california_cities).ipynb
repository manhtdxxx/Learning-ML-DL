```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_rows = 9999
df = pd.read_csv('california_cities.csv')
df
```

```python
df.info()
```

```python
df.area_total_km2.sort_values(ascending=False)
```

```python
lat = df.latd
lon = df.longd
population = df.population_total
area = df.area_total_km2

plt.figure(figsize=(8,6))
plt.scatter(lon,lat, c=np.log10(population), cmap='viridis', s=area, alpha=0.4 )
# c: chọn nhiều màu
# cmap: color map
# s: chọn nhiều size
# alpha: transparency

#
plt.axis('equal')
plt.title('Caliafornia Cities: Population & Area Distrbution')
plt.xlabel('LONGTITUDE')
plt.ylabel('LATITUDE')


# TẠO COLOR BAR CHO SCATTER
plt.colorbar(label='log$_{10}$(population)')  # hiển thị color map 
plt.clim(3,7) # colorbar limit


# TẠO LEGEND CHO SCATTER
area_range = [1300,900,500,300,100]
for area in area_range:
    plt.scatter([],[], s=area, c='k', alpha=0.5, 
                label=str(area)+' km$^2$')
    
plt.legend(loc=1, title='City Area', labelspacing=1)
```
