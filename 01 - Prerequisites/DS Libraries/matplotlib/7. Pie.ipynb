```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = ['Điểm Văn','Điểm Toán','Điểm Hóa','Điểm Lý']
y = [4,7,2,9]

fig1, sub1 = plt.subplots(figsize=(8,4))

sub1.pie(y,labels=x,colors=['yellow','green','blue','red'],autopct='%2.2f%%',explode=(0.1,0.1,0.1,0.1),shadow=True)

sub1.legend(loc=1,title='Score:')

sub1.set_title('Pie',fontsize = 30)
```
