```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = [0,1,2,3] ; 
x = np.array(x) # converted to array to make caculations
x1 = x - 0.1 # dịch trái 
x2 = x + 0.1 # dịch phải
y1 = [4,7,2,9]
y2 = [3,5,7,8]

fig1, sub1 = plt.subplots(nrows=1,ncols=1,figsize=(8,4))
sub1.bar(x1,y1,color='green',edgecolor='red',linewidth=2,width=0.2, label='Học sinh A')
sub1.bar(x2,y2,color='blue',edgecolor='yellow',linewidth=2,width=0.2, label='Học sinh B')
sub1.legend(loc=2)

sub1.set_xlim(-0.5,3.5)
sub1.set_ylim(0,10)

sub1.set_xticks([0,1,2,3])
sub1.set_xticklabels(['Điểm Văn','Điểm Toán','Điểm Hóa','Điểm Lý'])

sub1.yaxis.grid()

sub1.set_xlabel('Môn học')
sub1.set_ylabel('Điểm')

fig1.suptitle('SIDE BY SIDE BAR')

fig1.savefig('Side by side bar.png')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = ['Điểm Văn','Điểm Toán','Điểm Hóa','Điểm Lý']
y1 = [4,7,2,9]
y2 = [3,5,7,8]

fig1, sub1 = plt.subplots(nrows=1,ncols=1,figsize=(8,4))
sub1.bar(x,y1,color='green',edgecolor='red',linewidth=5,width=0.5, label='Học sinh A')
sub1.bar(x,y2,color='blue',edgecolor='yellow',linewidth=5,width=0.5, label='Học sinh B', bottom=y1)
sub1.legend(loc=2)

sub1.set_xlim(-0.5,3.5)
sub1.set_ylim(0,20)

sub1.yaxis.grid()
#sub1.xaxis.grid()

sub1.set_xlabel('Các môn học',fontsize=10)
sub1.set_ylabel('Điểm',fontsize=10)

fig1.suptitle('Stacked bar',fontsize=20)

fig1.savefig('Stacked bar')
```
