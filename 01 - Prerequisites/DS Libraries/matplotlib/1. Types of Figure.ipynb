# HIST

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a = ['Mạnh','Lộc','Huy','Quân','Dũng','Minh']
x = [2,8,3,7,9,2] # 2 xuất hiện 2 lần
y = [4,7,4,9,7,4] # 4 xuất hiện 3 lần, 7 xuất hiện 2 lần, 9 xuất hiện 1 lần

df = pd.DataFrame({'Họ tên':a,'Điểm Toán':x,'Điểm Văn':y})
print(df)


# histogram: biểu đồ tần suất
df.hist()
```

```python
rng = np.random.default_rng(1)
x = rng.normal(170,10,250) # mean = 170, standard deviation = 10, 250 values

plt.hist(x)
```

# PLOT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a = ['Mạnh','Lộc','Huy','Quân','Dũng','Minh']
x = [2,8,3,7,9,2] 
y = [4,7,4,9,7,4] 

df = pd.DataFrame({'Họ tên':a,'Điểm Toán':x,'Điểm Văn':y})
print(df)

# plot: biểu đồ đường
df.plot()
```

# BOXPLOT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a = ['Mạnh','Lộc','Huy','Quân','Dũng','Minh']
x = [2,8,3,7,9,2] 
y = [4,7,4,9,7,4] 

df = pd.DataFrame({'Họ tên':a,'Điểm Toán':x,'Điểm Văn':y})
print(df)

# boxplot: biểu đồ nến
df.boxplot()

#plt.close() # đã đóng và không thể show ra
```
