```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = ['Điểm Văn','Điểm Toán','Điểm Hóa','Điểm Lý']
y = [4,7,2,9]

fig1, sub1 = plt.subplots(nrows=1,ncols=1,figsize=(8,4))
sub1.bar(x,y,color='green',edgecolor='red',linewidth=5)

sub1.set_xlim(-1,4)
sub1.set_ylim(0,10)

sub1.yaxis.grid()
#sub1.xaxis.grid()

sub1.set_xlabel('Các môn học',fontsize=15)
sub1.set_ylabel('Điểm',fontsize=15)

fig1.suptitle('Vertical bar',fontsize=30)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = ['Điểm Văn','Điểm Toán','Điểm Hóa','Điểm Lý']
y = [4,7,2,9]

fig1, sub1 = plt.subplots(nrows=1,ncols=1,figsize=(8,4))
sub1.barh(x,y,color='green',edgecolor='red',linewidth=5)

sub1.set_ylim(-1,4)
sub1.set_xlim(0,10)

sub1.xaxis.grid()
#sub1.yaxis.grid()

sub1.set_ylabel('Các môn học',fontsize=15)
sub1.set_xlabel('Điểm',fontsize=15)

fig1.suptitle('Horizontal bar',fontsize=30)
```
